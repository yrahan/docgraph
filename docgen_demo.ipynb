Automatic Documentation Generator Prototype.============================================

Context presentation--------------------The idea behind this prototype is to study the feasability of an automatique mapping of the relation between :* scripts launching sql procedures.* sql procedures.* databases.The result will be a flow chart stored as a *dot file* and with a visualisation tool (Graphviz).A possible extention of our project, could be the add of an analyzing tools for performance and other developpement and production purpose.
